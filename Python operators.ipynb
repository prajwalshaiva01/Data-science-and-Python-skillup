 ## Python operation 

### Arithmetic operation

```python
a=10
b=30
print('add',a+b)
print('sub',a-b)
print('mul',a*b)
print('div',a/b)
print('floor div',a//b)
print('modulus',a%b)
print('exponentiation',a**b)
```

### Comparison Operators

```python
a=10
b=20
print(a<b)
print(a>b)
print(a==b)
print(a != b)
print(a>=b)
print(a<=b)
```

### Logical Operators

```python
a=True
b=False
print(a and b)
print(a or b)
print(not b)
print(not a)
```

### Bitwise Operators

```python
a=10
b=20
print(a&b)
print(a|b)
print(~b)
print(~a)
print(a^b)
print(a>>b)
print(a<<b)
```

### Assignment Operators

```python
a=10
b=a
print(b)
b+=a
print(b)
b-=a
print(b)
b*=a
print(b)
b/=a
print(b)
```

### Identity Operators


```python
a=10
b=20
c=a
print(a is not b)
print(a is b)
print(a is c)
```

### Mebership Operators


```python
a=10
b=20
list=[10,11,13,24,35]
if (a not in list):
  print('a is not present in given list')
else:
     print('a is present in given list')
if (b not in list):
  print('b is not present in given list')
else:
     print('b is present in given list')

```

### Ternary Operator

```python
a,b=10,20
num= a if a<b else b
print(num)
```
