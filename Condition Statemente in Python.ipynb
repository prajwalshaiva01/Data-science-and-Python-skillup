### if else Conditional Statement

```python
age=10
if age<=10:
    print('travel is free')
else:
    print('pay for ticket')
```

### elif Statement

```python
age=int(input())
if age<=10:
    print('child')
elif age>=20 and age>=30:
    print('teenger')
elif age>=60:
    print('youger adult')
else:
    print('adult')
```
